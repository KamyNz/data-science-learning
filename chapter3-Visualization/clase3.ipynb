## Scatter plot

```python
# Using matplolib
from matplotlib import pyplot as plt
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

print(years)
print(gdp)
```

```python
# create a line chart, years on x-axis, gdp on y-axis
plt.plot(years,gdp, color="green",marker="o",linestyle="solid")
# add title to plot
plt.title("Nominal GDP")
# add label to y-axis
plt.ylabel("Billions of $")
plt.show()
```

# Barchart

A bar chart is a good choice when you want to show how some quantity varies among some discrete
set of items.

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
# plot bars with left x-coordinates [0,1,2,3,4]; heights [num_oscars]
plt.bar(range(len(movies)),num_oscars)
plt.title("Favorites Movies")
plt.ylabel("# Of Academy Awards")

# label x-axis with movie namesa at bar centers
plt.xticks(range(len(movies)),movies)
#plt.xticks
plt.show()
```

A bar chart can also be a good choice for plotting histograms of bucketed numeric values, as in
Figure 3-3, in order to visually explore how the values are distributed

```python
from collections import Counter
```

```python
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
```

```python
# Bucket grades by decile, but put 100 in with the 90s

# with for loop to understand list compression
for grade in grades:
    minTemp = min(grade // 10*10,90)
    print(minTemp)

# List compression
list_obj = [min(grade // 10*10,90) for grade in grades]
counter_obj = Counter(list_obj) # Counter class transform list to dict
print(counter_obj)
```

```python
# Making histogram
plt.bar([x + 5 for x in counter_obj.keys()], # Shift bars right by 5
       counter_obj.values(),                 # Give each bar its correct height
       10,                                   # Give each bar a width of 10  
       edgecolor=(0,0,0))                    # Black edges for each bar

# Arranging x-axis and y-axis
plt.axis([-5,105,0,5])                       # x-axis from -5 to 105 and y-axis from 0 to 
plt.xticks([10 *i for i in range(11)])       # x-axis labels at 0,10, ... , 100

# Putting labels
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
```

```python
## OJO, When creating barchart (histogram) it is consider especially bad form for your y-axis not to start
## at 0, since this is an easy way to mislead people 

# Misleading Example
mentions = [500, 505]
years = [2017, 2018]
plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times I heard someone say 'data science'")
# if you don't do this, matplotlib will label the x-axis 0, 1
# and then add a +2.013e3 off in the corner (bad matplotlib!)
plt.ticklabel_format(useOffset=False)
# misleading y-axis only shows the part above 500
plt.axis([2016.5, 2018.5, 499, 506])
plt.title("Look at the 'Huge' Increase! | Misleading Example")
plt.show()
```

```python
## In comparison to using more sensible axes
mentions = [500, 505]
years = [2017, 2018]
plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times I heard someone say 'data science'")
# if you don't do this, matplotlib will label the x-axis 0, 1
# and then add a +2.013e3 off in the corner (bad matplotlib!)
plt.ticklabel_format(useOffset=False)
# More sensible axis
plt.axis([2016.5, 2018.5, 0, 550])
plt.title("Not So Huge Anymore")
plt.show()
```

## Line charts

As we saw already, we can make line charts using plt.plot. These are a good choice for showing
trends

```python
## Line Charts (Scatter plot but with lines)

# Making line charts
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]

# Using list compression and zip to get compression
total_error = [x + y for x, y in zip(variance,bias_squared)]
print("Presume total_error")
print(total_error)
print("\n")

# Getting xs, this object allows to make multiple calls to plt.plot
xs = [i for i, _ in enumerate(variance)] # putting _ to not take into account second element obtained in enumerate
print(xs) 

# We can multiple calls to plt.plot
# to show multiple series on the same chart
plt.plot(xs,variance, 'g-',label="variance")
plt.plot(xs,bias_squared,'r-.',label="bias^2")
plt.plot(xs,total_error,'b:',label="total error")

# g- means green solid line
# r-. dot-dashed line
# blue dotted line

# Because we've assigned labels to each series,
# we can get a legend for free (loc=9 means "top center")

plt.legend(loc=9) # Putting legend of colors in center
plt.xlabel("Model complexity") # x-label
plt.xticks([])
plt.title("Bias-Variance Tradeoff")
plt.show()
```
